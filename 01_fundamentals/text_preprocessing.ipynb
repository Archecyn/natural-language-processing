# Text Preprocessing

This notebook covers fundamental text preprocessing techniques including:
- **Tokenisation**: Splitting text into individual words or tokens
- **Stemming**: Reducing words to their root form by removing suffixes
- **Lemmatisation**: Converting words to their dictionary base form (more accurate than stemming)

These techniques form the foundation of natural language processing and are essential for preparing text data for machine learning models.

```python
# Import required libraries
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re

# Download required NLTK data resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('stopwords')

print("NLTK resources downloaded successfully.")
```

## 1. Tokenisation

Tokenisation is the process of breaking text into smaller units called tokens. These tokens are typically words, but can also be sentences or sub-word units.

NLTK provides two main types of tokenisation:
- **Word tokenisation**: Splits text into individual words
- **Sentence tokenisation**: Splits text into sentences

```python
# Sample text for demonstration
sample_text = "Natural language processing (NLP) is a subfield of linguistics, computer science, and artificial intelligence concerned with the interactions between computers and human language. It focuses on programming computers to process and analyse large amounts of natural language data."

# Word tokenisation - split text into individual words
word_tokens = word_tokenize(sample_text)
print("=== Word Tokenisation ===")
print(f"Original text: {sample_text[:100]}...")
print(f"\nNumber of tokens: {len(word_tokens)}")
print(f"\nTokens: {word_tokens}")
```

```python
# Sentence tokenisation - split text into sentences
sentence_tokens = sent_tokenize(sample_text)
print("=== Sentence Tokenisation ===")
print(f"Number of sentences: {len(sentence_tokens)}")
print(f"\nSentences:")
for i, sent in enumerate(sentence_tokens, 1):
    print(f"  {i}. {sent}")
```

## 2. Stemming

Stemming is a rule-based technique that removes suffixes from words to reduce them to their root form. The Porter Stemmer is one of the most commonly used stemming algorithms.

**Note**: Stemming may produce non-actual words (e.g., "running" → "run"), but it's useful for information retrieval where word variations have similar meanings.

```python
# Initialise the Porter Stemmer
stemmer = PorterStemmer()

# Example words to stem
words_to_stem = ["running", "runs", "runner", "easily", "happiness", "connected", "connecting", "connection", "studies", "studying"]

print("=== Stemming Examples ===")
print(f"{'Original Word':<15} {'Stemmed Form':<15}")
print("-" * 30)
for word in words_to_stem:
    stemmed = stemmer.stem(word)
    print(f"{word:<15} {stemmed:<15}")
```

```python
# Apply stemming to our tokenised text
stemmed_tokens = [stemmer.stem(token) for token in word_tokens if token.isalpha()]
print("=== Stemmed Tokens from Sample Text ===")
print(f"Original tokens (first 20): {word_tokens[:20]}")
print(f"\nStemmed tokens (first 20): {stemmed_tokens[:20]}")
```

## 3. Lemmatisation

Lemmatisation is a more sophisticated technique that converts words to their dictionary base form (lemma). Unlike stemming, lemmatisation produces actual words that can be found in a dictionary.

The WordNet Lemmatizer uses WordNet's database of word relationships to perform lemmatisation, which is more accurate than stemming but slower.

```python
# Initialise the WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Example words to lemmatise
words_to_lemmatise = ["running", "runs", "runner", "easily", "happiness", "connected", "connecting", "connection", "studies", "studying", "better", "best"]

print("=== Lemmatisation Examples ===")
print(f"{'Original Word':<15} {'Lemmatised Form':<15}")
print("-" * 30)
for word in words_to_lemmatise:
    lemmatised = lemmatizer.lemmatize(word)
    print(f"{word:<15} {lemmatised:<15}")
```

```python
# Lemmatisation with different POS tags
# By default, lemmatiser treats all words as nouns
# We can specify the POS tag for more accurate lemmatisation

print("=== Lemmatisation with POS Tags ===")
print(f"{'Word':<15} {'Noun (default)':<20} {'Verb':<20}")
print("-" * 55)

words_for_pos = ["running", "flies", "better", "studies"]
for word in words_for_pos:
    noun_form = lemmatizer.lemmatize(word)  # Default: noun
    verb_form = lemmatizer.lemmatize(word, pos='v')  # Verb form
    print(f"{word:<15} {noun_form:<20} {verb_form:<20}")
```

## 4. Stop Words Removal

Stop words are common words that carry little semantic meaning (e.g., "the", "is", "at", "which"). Removing them can reduce noise in the data and improve processing efficiency.

```python
# Get English stop words
stop_words = set(stopwords.words('english'))

print("=== English Stop Words ===")
print(f"Total number of stop words: {len(stop_words)}")
print(f"\nSample stop words: {list(stop_words)[:20]}")
```

```python
# Remove stop words from our tokenised text
filtered_tokens = [token for token in word_tokens if token.lower() not in stop_words and token.isalpha()]

print("=== After Stop Words Removal ===")
print(f"Original word count: {len([t for t in word_tokens if t.isalpha()])}")
print(f"Filtered word count: {len(filtered_tokens)}")
print(f"\nFiltered tokens: {filtered_tokens}")
```

## 5. Complete Text Preprocessing Pipeline

Let's combine all the techniques into a complete preprocessing pipeline that can be reused for any text data.

```python
def preprocess_text(text, use_stemming=False, remove_stopwords=True):
    """
    Complete text preprocessing pipeline.
    
    Parameters:
    -----------
    text : str
        Input text to preprocess
    use_stemming : bool
        Whether to apply stemming (True) or lemmatisation (False)
    remove_stopwords : bool
        Whether to remove stop words
    
    Returns:
    --------
    list
        List of preprocessed tokens
    """
    # Initialise processors
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    
    # Convert to lowercase
    text = text.lower()
    
    # Tokenise
    tokens = word_tokenize(text)
    
    # Remove non-alphabetic tokens
    tokens = [token for token in tokens if token.isalpha()]
    
    # Remove stop words
    if remove_stopwords:
        tokens = [token for token in tokens if token not in stop_words]
    
    # Apply stemming or lemmatisation
    if use_stemming:
        tokens = [stemmer.stem(token) for token in tokens]
    else:
        tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    return tokens

# Test the preprocessing pipeline
print("=== Complete Text Preprocessing Pipeline ===\n")

test_text = "The quick brown foxes were jumping over the lazy dogs while the birds were singing beautifully."

print(f"Original text: {test_text}\n")

# With lemmatisation (default)
processed = preprocess_text(test_text, use_stemming=False, remove_stopwords=True)
print(f"With lemmatisation: {processed}")

# With stemming
processed_stemmed = preprocess_text(test_text, use_stemming=True, remove_stopwords=True)
print(f"With stemming: {processed_stemmed}")

# Without stop words removal
processed_no_stop = preprocess_text(test_text, use_stemming=False, remove_stopwords=False)
print(f"Without stop words removal: {processed_no_stop}")
```

## 6. Practical Example: Analysing a Document

Let's apply our preprocessing pipeline to a real document and extract meaningful statistics.

```python
# Sample document
document = """
Natural Language Processing (NLP) is an exciting field of artificial intelligence 
that focuses on enabling computers to understand, interpret, and manipulate human language.
NLP combines computational linguistics with machine learning and deep learning techniques
to process and analyse large amounts of natural language data.

Applications of NLP include machine translation, sentiment analysis, chatbots,
text summarisation, named entity recognition, and question answering systems.
Recent advances in transformer models have significantly improved the accuracy
of NLP systems, making them more useful for real-world applications.
"""

# Preprocess the document
processed_doc = preprocess_text(document, use_stemming=False, remove_stopwords=True)

print("=== Document Analysis ===")
print(f"\nOriginal document length: {len(document)} characters")
print(f"Number of tokens after preprocessing: {len(processed_doc)}")
print(f"\nPreprocessed tokens: {processed_doc}")

# Calculate token frequency
from collections import Counter
token_freq = Counter(processed_doc)

print(f"\n=== Top 10 Most Common Words ===")
for word, count in token_freq.most_common(10):
    print(f"  {word}: {count}")
```
