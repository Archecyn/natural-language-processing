# Word2Vec

This notebook covers Word2Vec, a groundbreaking word embedding technique introduced by Google in 2013. Word2Vec represents words as dense vectors in a continuous vector space, capturing semantic relationships between words.

Unlike sparse representations like Bag of Words or TF-IDF, Word2Vec produces low-dimensional vectors where similar words are close together in the embedding space.

```python
# Import required libraries
import numpy as np
import pandas as pd
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

print("Libraries imported successfully.")
```

## 1. Understanding Word2Vec

Word2Vec uses neural networks to learn word embeddings from large corpora. The key idea is that words that appear in similar contexts should have similar meanings.

**Two architectures:**
- **CBOW (Continuous Bag of Words)**: Predicts a word from its surrounding context
- **Skip-gram**: Predicts surrounding words from a given word

Skip-gram works well with smaller datasets, while CBOW is faster and performs better with frequent words.

```python
# Sample corpus for demonstration
corpus = [
    "Natural language processing is a fascinating field of study",
    "Machine learning and deep learning are important for AI",
    "Neural networks can learn complex patterns from data",
    "Text classification is useful for sentiment analysis",
    "Word embeddings capture semantic meaning of words",
    "Transformer models have revolutionised NLP",
    "Deep learning enables computers to understand language",
    "Artificial intelligence is transforming many industries",
]

# Tokenise the corpus
tokenised_corpus = [doc.lower().split() for doc in corpus]

print("=== Sample Corpus ===")
for i, doc in enumerate(tokenised_corpus, 1):
    print(f"Document {i}: {doc}")
```

```python
# Train Word2Vec model using CBOW (default)
model_cbow = Word2Vec(
    sentences=tokenised_corpus,
    vector_size=100,      # Dimensionality of word vectors
    window=2,             # Context window size
    min_count=1,         # Minimum word frequency
    workers=4,           # Number of threads for training
    epochs=100,          # Number of training epochs
    sg=0                 # 0 for CBOW, 1 for Skip-gram
)

print("=== Word2Vec CBOW Model ===")
print(f"Vocabulary size: {len(model_cbow.wv)}")
print(f"Vector size: {model_cbow.wv.vector_size}")
print(f"\nVocabulary: {list(model_cbow.wv.key_to_index.keys())}")
```

```python
# Train Word2Vec model using Skip-gram
model_sg = Word2Vec(
    sentences=tokenised_corpus,
    vector_size=100,
    window=2,
    min_count=1,
    workers=4,
    epochs=100,
    sg=1                 # Skip-gram
)

print("=== Word2Vec Skip-gram Model ===")
print(f"Vocabulary size: {len(model_sg.wv)}")
print(f"Vector size: {model_sg.wv.vector_size}")
```

## 2. Word Vector Operations

One of the most powerful features of Word2Vec is that word vectors capture semantic relationships. This allows us to perform vector arithmetic to find analogies.

```python
# Get word vectors
print("=== Word Vector Examples ===\n")

sample_words = ["machine", "learning", "neural", "language"]
for word in sample_words:
    if word in model_cbow.wv:
        vector = model_cbow.wv[word]
        print(f"Word: '{word}'")
        print(f"  Vector shape: {vector.shape}")
        print(f"  First 5 values: {vector[:5]}")
        print()
```

```python
# Find most similar words
print("=== Most Similar Words ===\n")

test_words = ["machine", "neural", "language"]
for word in test_words:
    if word in model_cbow.wv:
        similar = model_cbow.wv.most_similar(word, topn=3)
        print(f"Words most similar to '{word}':")
        for sim_word, score in similar:
            print(f"  {sim_word}: {score:.4f}")
        print()
```

## 3. Vector Arithmetic (Word Analogies)

Word2Vec captures linear relationships between words. A famous example is:
- king - man + woman ≈ queen

This demonstrates that the model has learned gender relationships from the data.

```python
# Demonstrate vector arithmetic
# Note: This requires a larger corpus to learn meaningful relationships

print("=== Vector Arithmetic ===\n")

# Example: king - man + woman = ?
# This is a classic example, but may not work with small corpus

# Let's try with our small corpus
try:
    # Simple analogy: learning + deep = deep + learning (should be similar)
    result = model_cbow.wv.most_similar(
        positive=["learning", "deep"],
        negative=["machine"],
        topn=3
    )
    print("learning + deep - machine:")
    for word, score in result:
        print(f"  {word}: {score:.4f}")
except KeyError as e:
    print(f"Could not perform operation: {e}")
    print("\nNote: Word analogies require a larger training corpus.")
```

## 4. Visualising Word Embeddings

We can visualise word embeddings in 2D using dimensionality reduction techniques like PCA or t-SNE.

```python
# Get all word vectors
words = list(model_cbow.wv.key_to_index.keys())
vectors = np.array([model_cbow.wv[word] for word in words])

print("=== Word Embeddings ===")
print(f"Number of words: {len(words)}")
print(f"Vector dimensions: {vectors.shape}")
```

```python
# Reduce to 2D using PCA
pca = PCA(n_components=2)
vectors_2d = pca.fit_transform(vectors)

print("=== PCA Reduction ===")
print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
print(f"Total variance explained: {sum(pca.explained_variance_ratio_):.4f}")
```

```python
# Plot the word embeddings
plt.figure(figsize=(12, 8))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], alpha=0.5)

# Add labels
for i, word in enumerate(words):
    plt.annotate(word, (vectors_2d[i, 0], vectors_2d[i, 1]), fontsize=10)

plt.title('Word2Vec Embeddings (PCA)', fontsize=14)
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 5. Training on Larger Data

Word2Vec works best with larger corpora. Let's train on a more substantial dataset using gensim's built-in corpus.

```python
# Use a larger corpus - we'll create sentences from a larger text
larger_corpus = [
    "The quick brown fox jumps over the lazy dog",
    "A fast fox can jump over a sleeping dog",
    "The lazy dog sleeps in the sun",
    "Quick foxes and lazy dogs are common in fables",
    "Machine learning is a subset of artificial intelligence",
    "Deep learning is a subset of machine learning",
    "Neural networks are used in deep learning",
    "Natural language processing uses neural networks",
    "Text classification is an important NLP task",
    "Sentiment analysis classifies text as positive or negative",
    "Named entity recognition identifies entities in text",
    "Machine translation translates text between languages",
    "Question answering systems answer questions from text",
    "Text summarisation creates summaries of documents",
    "Information retrieval finds relevant documents",
    "Search engines use information retrieval techniques",
    "Word embeddings capture semantic meaning",
    "Word2Vec creates word embeddings from text",
    "GloVe is another word embedding technique",
    "BERT is a transformer-based language model",
]

tokenised_larger = [doc.lower().split() for doc in larger_corpus]

# Train a new model
model_large = Word2Vec(
    sentences=tokenised_larger,
    vector_size=50,
    window=3,
    min_count=1,
    epochs=200,
)

print("=== Larger Corpus Word2Vec Model ===")
print(f"Vocabulary size: {len(model_large.wv)}")
```

```python
# Test similarity on larger model
print("=== Similarity Tests on Larger Model ===\n")

test_pairs = [
    ("machine", "learning"),
    ("deep", "learning"),
    ("neural", "networks"),
    ("text", "classification"),
]

for word1, word2 in test_pairs:
    if word1 in model_large.wv and word2 in model_large.wv:
        similarity = model_large.wv.similarity(word1, word2)
        print(f"Similarity('{word1}', '{word2}'): {similarity:.4f}")
```

## 6. Saving and Loading Models

Word2Vec models can be saved and loaded for later use.

```python
# Save the model
model_large.save("word2vec_model.bin")

# Load the model
loaded_model = Word2Vec.load("word2vec_model.bin")

print("=== Model Save/Load ===")
print(f"Original model vocabulary: {len(model_large.wv)}")
print(f"Loaded model vocabulary: {len(loaded_model.wv)}")

# Test similarity with loaded model
sim = loaded_model.wv.similarity("machine", "learning")
print(f"\nSimilarity test after loading: {sim:.4f}")
```

```python
import os
# Clean up saved model
if os.path.exists("word2vec_model.bin"):
    os.remove("word2vec_model.bin")
    print("Cleaned up saved model file.")
```
